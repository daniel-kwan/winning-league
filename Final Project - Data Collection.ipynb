```python
# Importing top North American players
with open("NA_Challenger.txt", encoding="utf8") as json_file:
    data = json.load(json_file)
    json_file.close()
```

```python
# Create empty dataframe
naChal = pd.DataFrame(columns=['summonerName', 'summonerID', 'puuid'])

# Populate with JSON data
for player in data['entries']:
    #row = {'summonerName' : player['summonerName'], 'summonerId' : player['summonerId'], 'puuid' : None}
    naChal.loc[len(naChal.index)] = [player['summonerName'], player['summonerId'], None] 

    # naChal.append(row, ignore_index = True)

# Get PUUID (unique identifier) for each player
for i in range(len(naChal)):
    lookup = naChal.iloc[i]['summonerName']
    try:
        naChal.iloc[i]['puuid'] = lw.summoner.by_name('na1', lookup)['puuid']
    except:
        pass
```

```python
# Failed PUUIDs can be dropped
for i in range(len(naChal)):
    if naChal.iloc[i]['puuid'] == None:
        print(i)
```

```python
matchIDs = set()

for i in range(len(naChal)):
    # Get match IDs for past 50 games, per player identified using puuid
    try:
        # print("Row", i)
        matchlist = lw.match.matchlist_by_puuid('na1', naChal.iloc[i]['puuid'], count=74, type="ranked")
        for match in matchlist:
            matchIDs.add(match)
    except:
        pass
    
matchIDs = list(matchIDs)
```

```python
def getData(matchList, region, iterations, batchNum, fileName):
    data = pd.DataFrame(columns=[
        'matchID',
        'matchDuration',
        'win',
        'champID', 
        'champName', 
        'position',
        'lane',
        'role',
        'side',
        'dragons',
        'heralds',
        'barons',
        'towers',
        'kills',
        'deaths',
        'assists',
        'vision',
        'gold',
        'level',
        'minions',
        'objDamage',
        'turretDamage',
        'champDamage',
        'takenDamage',
        'oppGold',
        'oppLevel',
        'oppMinions',
        'gold@10',
        'level@10',
        'minions@10',
        'oppGold@10',
        'oppLevel@10',
        'oppMinions@10'
    ])
    
    startTime = time.time()
    skips = []
    
    for i in range(len(matchList)):
        print(matchList[i])
        
        
        
        # Get match & match timeline by match ID
        try: 
            match = lw.match.by_id(region, matchList[i])
            timeline = lw.match.timeline_by_match(region, matchList[i])

            # Select random player and identify opponent
            player = random.randint(0,9)
            opp = None
            if player <= 4:
                opp = player + 5
            else:
                opp = player - 5

            # print(player)
            # print(opp)

            row = []

            matchTime = match['info']['gameDuration']

            if matchTime > 300:

                # Match ID
                row.append(matchList[i])

                # Match Duration
                row.append(matchTime)

                # Win/Loss
                row.append(match['info']['participants'][player]['win'])

                # Champ ID
                row.append(match['info']['participants'][player]['championId'])

                # Champ Name
                row.append(match['info']['participants'][player]['championName'])

                # Position
                row.append(match['info']['participants'][player]['individualPosition'])

                # Lane
                row.append(match['info']['participants'][player]['lane'])

                # Role
                row.append(match['info']['participants'][player]['role'])

                # Side, plus all objectives
                if match['info']['participants'][player]['teamId'] == 100:
                    row.append('Blue')
                    row.append(match['info']['teams'][0]['objectives']['dragon']['kills'])
                    row.append(match['info']['teams'][0]['objectives']['riftHerald']['kills'])
                    row.append(match['info']['teams'][0]['objectives']['baron']['kills'])
                    row.append(match['info']['teams'][0]['objectives']['tower']['kills'])
                else:
                    row.append('Red')
                    row.append(match['info']['teams'][1]['objectives']['dragon']['kills'])
                    row.append(match['info']['teams'][1]['objectives']['riftHerald']['kills'])
                    row.append(match['info']['teams'][1]['objectives']['baron']['kills'])
                    row.append(match['info']['teams'][1]['objectives']['tower']['kills'])

                # Kills
                row.append(match['info']['participants'][player]['kills'])

                # Deaths
                row.append(match['info']['participants'][player]['deaths'])

                # Assists
                row.append(match['info']['participants'][player]['assists'])

                # Vision Score
                row.append(match['info']['participants'][player]['visionScore'])

                # Final Gold
                row.append(match['info']['participants'][player]['goldEarned'])

                # Final Level
                row.append(match['info']['participants'][player]['champLevel'])

                # Final CS
                row.append(match['info']['participants'][player]['totalMinionsKilled'])

                # Objective Damage
                row.append(match['info']['participants'][player]['damageDealtToObjectives'])

                # Turret Damage
                row.append(match['info']['participants'][player]['damageDealtToTurrets'])

                # Champ Damage
                row.append(match['info']['participants'][player]['totalDamageDealtToChampions'])

                # Damage Taken
                row.append(match['info']['participants'][player]['totalDamageTaken'])

                # Opponent Final Gold
                row.append(match['info']['participants'][opp]['goldEarned'])

                # Opponent Final Level
                row.append(match['info']['participants'][opp]['champLevel'])

                # Opponent Final CS
                row.append(match['info']['participants'][opp]['totalMinionsKilled'])



                if matchTime > 610:

                    # Gold at 10 minutes
                    row.append(timeline['info']['frames'][11]['participantFrames'][str(player+1)]['totalGold'])

                    # Level at 10 minutes
                    row.append(timeline['info']['frames'][11]['participantFrames'][str(player+1)]['level'])   

                    # CS at 10 minutes
                    row.append(timeline['info']['frames'][11]['participantFrames'][str(player+1)]['minionsKilled'] + 
                               timeline['info']['frames'][11]['participantFrames'][str(player+1)]['jungleMinionsKilled'])

                    # Opponent Gold at 10 minutes
                    row.append(timeline['info']['frames'][11]['participantFrames'][str(opp+1)]['totalGold'])

                    # Opponent Level at 10 minutes
                    row.append(timeline['info']['frames'][11]['participantFrames'][str(opp+1)]['level'])

                    # Opponent CS at 10 minutes
                    row.append(timeline['info']['frames'][11]['participantFrames'][str(opp+1)]['minionsKilled'] +
                               timeline['info']['frames'][11]['participantFrames'][str(opp+1)]['jungleMinionsKilled'])

                else:
                    for i in range(0,6):
                        row.append(None)

                # Add to dataframe
                data.loc[len(data.index)] = row

                print(f"Match complete. Time elapsed: {time.time() - startTime}")

            else:
                print(f"Match {matchList[i]} skipped - Duration")

    #         if time.time() - startTime > iterations:
    #                 print(f"Completed up to index {i}. Total time taken: {time.time() - startTime}")
    #                 data.to_csv(f"naMatchData{batchNum}.csv", index=False)
    #                 return data
                
        except:
            print(f"Match {matchList[i]} skipped - Timeout / Other")
            skips.append(matchList[i])

    
    data.to_csv(f"{fileName}{batchNum}.csv", index=False)
    print(f"Data completed. Total time taken: {time.time() - startTime}")
    print(f"{len(skips)} matches failed due to timeouts.")
    return skips
```
